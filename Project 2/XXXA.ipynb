## Task 15: Get the Metrics of the Trained $M_0$ kNN Model

```python
##########################################
### DO NOT EDIT / DELETE THIS FUNCTION ###
##########################################
import sklearn.metrics as metrics
import numpy as np

def display_metrics(y_true, y_pred):
    accuracy = metrics.accuracy_score(y_true, y_pred)
    bal_accuracy = metrics.balanced_accuracy_score(y_true, y_pred)
    precision = metrics.precision_score(y_true, y_pred, average = 'weighted')
    
    f1 = metrics.f1_score(y_true, y_pred, average = 'weighted')
    fmi = metrics.fowlkes_mallows_score(y_true, y_pred)

    print(f'Accuracy              = {accuracy:,.4f}')
    print(f'Balanced Accuracy     = {bal_accuracy:,.4f}')
    print(f'Precision             = {precision:,.4f}')
    print(f'F1-Score              = {f1:,.4f}')
    print(f'Fowlkes-Mallows Index = {fmi:,.4f}')

    return None
##########################################
### DO NOT EDIT / DELETE THIS FUNCTION ###
##########################################
```
